```python
#converting negative values to positive
app_data[["DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION","DAYS_ID_PUBLISH","DAYS_LAST_PHONE_CHANGE"]] = app_data[["DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION","DAYS_ID_PUBLISH","DAYS_LAST_PHONE_CHANGE"]].apply(abs)
```

```python
#converting dtype of columns to 'object' 
obj = app_data.columns[app_data.nunique() <= 50]
app_data[obj] = app_data[obj].astype(object)

#creating new field 
app_data['Client_Age'] = app_data['DAYS_BIRTH'].apply(lambda x: x//365)
app_data['Employed_Years'] = app_data['DAYS_EMPLOYED'].apply(lambda x: x//365)
```

```python
app_data["SK_ID_CURR"] = app_data["SK_ID_CURR"].astype(object) 
app_data.loc[app_data["CODE_GENDER"] == "XNA","CODE_GENDER"] = np.NaN
app_data.describe()
```

```python
#dropping null values
app_data = app_data[app_data['DAYS_LAST_PHONE_CHANGE'].notna()]
app_data = app_data[app_data['CNT_FAM_MEMBERS'].notna()]
app_data = app_data[app_data['CODE_GENDER'].notna()]
app_data = app_data[app_data['AMT_ANNUITY'].notna()]
app_data = app_data[app_data['AMT_GOODS_PRICE'].notna()]
app_data = app_data[app_data['EXT_SOURCE_2'].notna()]

app_data.isnull().sum()
```

# conclusion

* Banks should concentrate less on income kinds like "Working," which have a greater rate of failed payments, and more on contract types like "Student," "Pensioner," and "Businessman" for lucrative company. The organisation will suffer less financial loss as a result.

* Loans to clients between the ages of 17 and 35 should be avoided, and loans to widows should be favoured. Loan applicants whose applications have previously been denied should also be avoided. The 4 to 6 lac loan amount range is the riskiest and should be avoided.

* Cash loans are not the best choice because they carry greater risk, but revolving loans do not. Males should not be chosen above females. Clients with less than a high school diploma should be avoided.

* For effective payments, banks should concentrate on contract kinds like "student," "pensioner," and "businessman," with dwelling types other than "co-op apartments." 'Repair' loans have a higher percentage of late payments that are unsuccessful. The least dangerous housing type is "With parents," thus banks should make an effort to draw more customers from this group.

* These findings generally imply that banks should concentrate on customers who pose less of a risk in terms of their background and financial history. Banks can minimise financial losses and increase profits by avoiding riskier customers and loan types.
